```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Laptop_price_prediction_dataset.csv")
df
```

```python
df['Brand'].value_counts().plot(kind='bar')
plt.title('Count of Laptops per Brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
avg_price_over_time = df.groupby('Date')['Price_USD'].mean()
plt.plot(avg_price_over_time.index, avg_price_over_time.values)
plt.title('Average Laptop Price Over Time')
plt.xlabel('Date')
plt.ylabel('Average Price (USD)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['RAM_GB'], df['Price_USD'])
plt.title('RAM Size vs Price')
plt.xlabel('RAM (GB)')
plt.ylabel('Price (USD)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Price_USD'], bins=20, edgecolor='k', alpha=0.7)
plt.title('Distribution of Laptop Prices')
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df.boxplot(column='Price_USD', by='Brand', grid=False)
plt.title('Price Distribution per Brand')
plt.xlabel('Brand')
plt.ylabel('Price (USD)')
plt.suptitle('') 
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
brand_counts = df['Brand'].value_counts()
plt.pie(brand_counts, labels=brand_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Market Share of Each Brand')
plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df['Date'] = pd.to_datetime(df['Date'])

df['Price_USD'] = pd.to_numeric(df['Price_USD'], errors='coerce')

df.dropna(subset=['Date', 'Price_USD'], inplace=True)
plt.figure(figsize=(10, 8))
numerical_features = ['RAM_GB', 'Storage_GB', 'Price_USD']
corr = df[numerical_features].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Area map
plt.figure(figsize=(14, 8))
df_sorted = df.sort_values(by='Date')
plt.fill_between(df_sorted['Date'], df_sorted['Price_USD'], color='skyblue', alpha=0.4)
plt.plot(df_sorted['Date'], df_sorted['Price_USD'], color='Slateblue', alpha=0.6, linewidth=2)
plt.title('Price Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.show()
```
