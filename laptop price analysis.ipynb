```python

df = pd.DataFrame({
    'A': [5, 2, 9, 1, 7],
    'B': [4, 8, 2, 6, 3],
    'C': [7, 3, 5, 9, 1]
})

print("Original DataFrame:")
print(df)

df_nsmallest_A = df.nsmallest(3, 'A')
print("\n3 smallest values in column 'A':")
print(df_nsmallest_A)

df_nsmallest_AB = df.nsmallest(2, ['A', 'B',])
print("\n2 smallest values in columns 'A' and 'B':")
print(df_nsmallest_AB)

df_nsmallest_ABC = df.nsmallest(3, ['A', 'B','C'])
print("\n3 smallest values in columns 'A' and 'B' and 'C':")
print(df_nsmallest_ABC)

df_nsmallest_keep_first = df.nsmallest(2, 'B', keep='first')
print("\n2 smallest values in column 'A' (keep='first'):")
print(df_nsmallest_keep_first)

df_nsmallest_keep_all = df.nsmallest(2, 'B', keep='all')
print("\n2 smallest values in column 'A' (keep='all'):")
print(df_nsmallest_keep_all)

df_nsmallest_keep_last = df.nsmallest(2, 'B', keep='last')
print("\n2 smallest values in column 'A' (keep='last'):")
print(df_nsmallest_keep_last)
df
```

```python
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({
    'Category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Value': [10, 20, 10, 30, 20, 10, 40, 20, 10],
    'Other': [1, 2, 3, 4, 5, 6, 7, 8, 9]
})

print("Original DataFrame:")
print(df)
```

```python
result = df.groupby('Category').apply(lambda x: x.nsmallest(2, 'Value'))
print("\n2 smallest values per group in column 'Value':")
print(result)
```
